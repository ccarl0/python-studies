```python
import matplotlib.pyplot as plt
import pandas
import numpy as np
```

```python
a = np.array([x for x in range(10)])

plt.plot(a)
plt.show()
```

```python
a = np.array([x for x in range(10)])
b = np.array([x**2 for x in a])
plt.plot(a)
plt.plot(b)
#plt.plot(a,b)
plt.show()
```

```python
plt.plot(a, color = "green")
plt.plot(b, color = "blue", lw = 10 )
plt.show()
```

```python
c = np.array([x/2 for x in b])
```

```python
plt.plot(a)
plt.plot(b)
plt.plot(c)
```

```python
plt.plot(a,b,c)
plt.plot(b,a,c)
plt.plot(c,a,b)
plt.plot(c,b,a)
plt.show()
```

```python
plt.plot(a, marker = "1", alpha = 0.3, label = "AAA")
plt.plot(b, ls = "-.", label = ":)")
plt.plot(c, ls = "--", label = "last one")
plt.title("Hello Wrld", fontsize = 20, fontweight = "bold", color = "White")
plt.legend(loc = "upper center")
plt.xticks(range(0,3),("One","String","Hi"))
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```
